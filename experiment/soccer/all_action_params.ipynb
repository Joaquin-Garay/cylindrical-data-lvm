```python
from cyl_lvm.distributions import MultivariateGaussian
%load_ext autoreload
%autoreload 2
import cyl_lvm as spr
import numpy as np
```

```python
mom = spr.TwoLayerMoM(
    layer1_mixture = spr.MixtureModel([spr.MultivariateGaussian() for _ in range(3)],
                                      init="k-means"),
    layer2_mixtures = [spr.MixtureModel([spr.VonMises() for _ in range(2)],
                                        init="k-means")
                       for _ in range(3)]
)
mom.layer1_mixture.set_params(
    component_params=[
        (np.array([10, 22]), np.array([[50, 0], [0, 50]])),
        (np.array([28, 13]), np.array([[60, 0.0], [0.0, 60]])),
        (np.array([38, 7]), np.array([[80, 0], [0, 40]])),
    ],
    weights=np.array([1 for _ in range(3)]),
)

mom.layer2_mixtures[0].set_params(
        component_params=[
        (0.0, 2.0),   # component 0
        (2.2, 2.0),   # component 1
        ],
        weights=np.array([1, 1]),
        )

mom.layer2_mixtures[1].set_params(
        component_params=[
        (0.5, 2.0),   # component 0
        (2.5, 2.0),   # component 1
        ],
        weights=np.array([1, 1]),
        )

mom.layer2_mixtures[2].set_params(
        component_params=[
        (0.5, 2.0),   # component 0
        (2.5, 2.0),   # component 1
        ],
        weights=np.array([1, 1]),
        )



mom.plot()
```
